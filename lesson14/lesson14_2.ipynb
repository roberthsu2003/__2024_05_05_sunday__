```python
import matplotlib.pyplot as plt

index = [3, 4, 5, 6, 7]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5000, 6000, 7000, 8000, 9000]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.bar(index, values)
axes.set_xticks(index,labels=subject)

for x,y in zip(index,values):
    axes.annotate(str(y),(x-0.2,y+100))

xmin, xmax, ymin, ymax = axes.axis()
xmin -= 0.2
xmax += 0.2
ymin = 3000
ymax += 1000
axes.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)
plt.show()
```

```python
import matplotlib.pyplot as plt

index = [3, 4, 5, 6, 7]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5127, 7865, 7123, 9321, 8456]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(index, values,'or-.')
axes.set_xticks(index,labels=subject)

for x,y in zip(index,values):
    axes.annotate(str(y),(x-0.2,y+100))

xmin, xmax, ymin, ymax = axes.axis()
xmin -= 0.2
xmax += 0.2
ymin = 3000
ymax += 1000
axes.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)
plt.show()
```

```python
import matplotlib.pyplot as plt
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}
subjects = list(student1.keys())
value1 = list(student1.values())
value2 = list(student2.values())

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,value1,'bo--')
axes.plot(subjects,value2,'ro-')
plt.show()
```
